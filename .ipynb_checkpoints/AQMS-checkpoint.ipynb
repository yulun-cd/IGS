```python
import pandas as pd
import geopandas as gpd
```

```python
import glob
csv_files=glob.glob('data/AQMS'+'/*.csv')

df=pd.concat((pd.read_csv(f) for f in csv_files))
df.head()
```

```python
df.shape
```

```python
df=df.drop(['Species','Units','Provisional or Ratified'],axis=1)
```

```python
df.head()
```

```python
df.groupby('Site').describe()
```

```python
valid_AQMS=df.dropna()['Site'].unique().tolist()
valid_AQMS
```

```python
df['valid']=df['Site'].isin(valid_AQMS)
```

```python
clean_df=df[df['valid']]
```

```python
clean_df
```

```python
len(clean_df['Site'].unique())
```

```python
clean_df=clean_df.drop('valid',axis=1)
```

```python
gdf=gpd.read_file('data/AQMS/AQMS.gpkg')
gdf
```

```python
def comparelist(list1,list2):
    opt=[]
    for elem in list1:
        if elem not in list2:
            opt.append(elem)
    return opt

missing=comparelist(valid_AQMS,gdf.iloc[gdf['siteid'].isin(valid_AQMS).tolist()]['siteid'].tolist())
missing
```
