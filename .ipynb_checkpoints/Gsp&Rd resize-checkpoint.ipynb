```python
import pandas as pd
import geopandas as gpd
```

```python
GB_Gsp=gpd.read_file('data/OS Open Greenspace (ESRI Shape File) GB/data/GB_GreenspaceSite.shp')
GB_Gsp.shape
```

```python
LD_wards=gpd.read_file("data/LD_boundary/London-wards-2018_ESRI/London_Ward_CityMerged.shp")
LD_wards.head()
```

```python
LD_wards['merge_on']='London'
LD_bd=LD_wards.dissolve(by='merge_on')
LD_bd
```

```python
LD_bd=LD_bd.drop(['NAME','GSS_CODE','DISTRICT','LAGSSCODE','HECTARES','NONLD_AREA'],axis=1)
LD_bd
```

```python
london=LD_bd.loc['London','geometry']
```

```python
GB_Gsp['Is_Within_LD']=GB_Gsp['geometry'].within(london)
```

```python
GB_Gsp[GB_Gsp['Is_Within_LD']].plot()
```

```python
from pathlib import Path

folder=Path('data/oproad_essh_gb/data')

Rd=pd.concat([gpd.read_file(shp) for shp in folder.glob('*.shp')]).pipe(gpd.GeoDataFrame)

Rd
```

```python
Rd['Is_Within_LD']=pd.concat([Rd.loc[i,'geometry'].intersection(london) for i in range(Rd.shape[0])])
```
