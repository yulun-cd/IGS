```python
# Read in data

BP = pd.read_csv('data/con_var/BP.csv')
Tmp = pd.read_csv('data/con_var/TEM.csv')
RH = pd.read_csv('data/con_var/RH.csv')
```

```python
def trans(df):
    new_df = pd.DataFrame(df.loc[:8759, 'ReadingDateTime'])
    for site in df['Site'].unique():
        new_df[site] = df[df['Site']==site].Value.values
    return new_df
```

```python
BP = trans(BP)
BP.describe()
```

```python
BP.drop(['GR4', 'GB6'], axis=1, inplace=True)
BP.plot(x='ReadingDateTime')
```

```python
Tmp = trans(Tmp)
Tmp.describe()
```

The minimum value of **HK6** is very abnormal

```python
Tmp.plot(x='ReadingDateTime', linewidth=0.2)
```

```python
Tmp.sort_values(by='HK6', ascending=True).head(50)
```

The values of HK6 seems to be very different from other sites, so the data might be problematic and hence removing it would be the best for the result.

```python
Tmp.drop('HK6', axis=1, inplace=True)
```

```python
Tmp.drop('RG1', axis=1, inplace=True)
```

```python
Tmp.plot(x='ReadingDateTime', linewidth=0.2)
```

```python
RH = trans(RH)
RH.describe()
```

```python
RH.drop(['BG1', 'GR4', 'HK6','RG1'], axis=1, inplace=True)
RH.plot(x='ReadingDateTime', linewidth=0.1)
```

```python
def fill_nan(df):
    for col in df.columns[1:]:
        val = df[col].values.copy()
        for i in range(len(val)):
            if np.isnan(val[i]):
                j = 1
                while (np.isnan(val[i+j])) & (j < 12):
                    j += 1
                if j==12:
                    for z in range(j+1):
                        val[i+z] = val[i+z-24]
                val[i] = val[i-1] + (val[i+j] - val[i-1]) / (j+1)
        df[col] = val
```

```python
fill_nan(BP)
```

```python
BP.info()
```

```python
fill_nan(Tmp)
Tmp.info()
```

```python
fill_nan(RH)
RH.info()
```

```python
cond = pd.DataFrame(BP['ReadingDateTime'])
cond['bp_mean'] = BP[BP.columns[1:]].mean(axis=1)
cond['tmp_mean'] = Tmp[Tmp.columns[1:]].mean(axis=1)
cond['rh_mean'] = RH[RH.columns[1:]].mean(axis=1)
```
